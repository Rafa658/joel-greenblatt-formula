```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('statusinvest-busca-avancada.csv', sep=';', decimal = ',')
df
```

```python
df = df[['TICKER', 'PRECO',
        'DY', 
        'P/VP',
        'EV/EBIT',
        'ROIC',
        ' LIQUIDEZ MEDIA DIARIA',
        ' VALOR DE MERCADO']]
df
```

```python
df = df.dropna()
```

```python
for col in df.columns:
    if df[col].dtype == object:
        df[col] = df[col].str.replace('.', '')
df
```

```python
# for col in df.columns[1:]:
#     if df[col].dtype == object:
#         # df[col] = df[col].str.replace(',', '.').astype(float)
#         df[col] = df[col].str.replace(',', '.').astype(float)

for col in df.columns[1:]:
    if df[col].dtype == object:
        df[col] = df[col].str.replace(',', '.').astype(float)

df
```

```python
for col in df.columns:
    print(df[col].dtype)
```

```python
df[['ROIC', 'EV/EBIT', 'P/VP', ' LIQUIDEZ MEDIA DIARIA']] = df[['ROIC', 'EV/EBIT', 'P/VP', ' LIQUIDEZ MEDIA DIARIA']].astype(float)
```

```python
df.dtypes
```

```python
for col in df.columns:
    if df[col].dtype != object:
        df = df[df[col] > 0]
df = df[df[' LIQUIDEZ MEDIA DIARIA'] > 150000]
```

```python
df
```

```python
df['EBIT/EV'] = 1 / df['EV/EBIT'] 
df = df.drop(['EV/EBIT'], axis = 1)
```

```python
df
```

```python
df['EBIT/EV Ranking'] = df['EBIT/EV'].rank(ascending= 0)
df['ROIC Ranking'] = df['ROIC'].rank(ascending = 0)
df['General'] = df['EBIT/EV Ranking'] + df['ROIC Ranking']
df
```

```python
df = df.sort_values(by = ['General'])

df.head(40)
```

```python
df.to_csv('export_dataframe.csv', header = True, index = False)
```
